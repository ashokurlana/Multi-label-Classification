```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Summary_evaluation/Relevance_final.csv")
df['list'] = df[df.columns[1:]].values.tolist()
new_df = df[['text', 'list']].copy()
new_df.head()
```

```python
train_size = 0.8
train_dataset=new_df.sample(frac=train_size,random_state=200)
test_dataset=new_df.drop(train_dataset.index).reset_index(drop=True)
train_dataset = train_dataset.reset_index(drop=True)
```

```python
import pandas as pd

from simpletransformers.classification import MultiLabelClassificationModel
from simpletransformers.classification import ClassificationModel, ClassificationArgs


# Train and Evaluation data needs to be in a Pandas Dataframe containing at least two columns, a 'text' and a 'labels' column. The `labels` column should contain multi-hot encoded lists.
# train_data = [
#     ["Example sentence 1 for multilabel classification.", [1, 1, 1, 1, 0, 1]]
# ] + [["This is another example sentence. ", [0, 1, 1, 0, 0, 0]]]
train_df = pd.DataFrame(train_dataset, columns=["text", "list"])

# eval_data = [
#     ["Example eval sentence for multilabel classification.", [1, 1, 1, 1, 0, 1]],
#     ["Example eval senntence belonging to class 2", [0, 1, 1, 0, 0, 0]],
# ]
eval_df = pd.DataFrame(test_dataset)

# You can use anyone of the below two models
# BERT ("bert", "bert-base-multilingual-cased")
# XLM ("xlm", xlm-mlm-100-1280)

# Create a MultiLabelClassificationModel
model = MultiLabelClassificationModel("xlm", "xlm-mlm-100-1280", num_labels=6,
    args={
        "reprocess_input_data": True,
        "overwrite_output_dir": True,
        "num_train_epochs": 3,
        "learning_rate":1e-4,
        "train_batch_size":1,
        "eval_batch_size":1,
        "evaluate_during_training_steps": 10000
    },
)
# You can set class weights by using the optional weight argument
print(train_df.head())

# Train the model
model.train_model(train_df)

# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(eval_df)
print(result)
print(model_outputs)

predictions, raw_outputs = model.predict(
    ["This thing is entirely different from the other thing. "]
)
print(predictions)
print(raw_outputs)
```
